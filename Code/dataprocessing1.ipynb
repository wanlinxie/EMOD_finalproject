```python
import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
def getdata(company,sources):
    stock = pd.read_csv("StockData/" + company + "_pred"+ '.csv')
    dates = stock.iloc[:,0].values
    #labels = stock.iloc[:,8].values
    diff = stock.iloc[:,7].values
    dates_label = {}
    for i, date in enumerate(dates):
        if '/' in date:
            replace = datetime.strptime(date,"%m/%d/%y")
            date = datetime.strftime(replace, "%Y-%m-%d")
        dates_label[date] = diff[i]
    news = pd.read_csv("Prediction/" + company + "_" + sources + "_prediction.csv",encoding='latin-1')
    result = {} # date, content_score
    result_file = "result/" + company + "_" + sources + "_prediction.csv"
    for _, row in news.iterrows():
        if '/' in row['dates']:
            replace = datetime.strptime(row['dates'],"%m/%d/%y")
            dd = datetime.strftime(replace, "%Y-%m-%d")
        else:
            dd = row['dates']
        result[dd] = result.get(dd,[]) + [row['content_scores']]
    final = {}
    d = list(result.keys())
    d.sort()
    for key in d:
        if key in dates_label:
            result[key] = np.mean(result[key])
            final['Dates'] = final.get('Dates', []) + [key]
            final['content_scores'] = final.get('content_scores',[]) + [result[key]]


    result = {}
    for _, row in news.iterrows():
        if '/' in row['dates']:
            replace = datetime.strptime(row['dates'],"%m/%d/%y")
            dd = datetime.strftime(replace, "%Y-%m-%d")
        else:
            dd = row['dates']
        result[dd] = result.get(dd,[]) + [row['title_score']]
    for key in final['Dates']:
        final['title_scores'] = final.get('title_scores',[]) + [np.mean(result[key])]
    result = {}
    for _, row in news.iterrows():
        if '/' in row['dates']:
            replace = datetime.strptime(row['dates'],"%m/%d/%y")
            dd = datetime.strftime(replace, "%Y-%m-%d")
        else:
            dd = row['dates']
        result[dd] = result.get(dd,[]) + [row['firstP_score']]
    for key in final['Dates']:
        final['firstP_scores'] = final.get('firstP_scores',[]) + [np.mean(result[key])]
        #final['stock_label'] = final.get('stock_label',[]) + [dates_label[key][0]]
        final['diff'] = final.get('diff',[]) + [dates_label[key]]

    data = pd.DataFrame.from_dict(final)
    data.to_csv(result_file)
```

```python
companies = ["NFLX", "FB","BA", "TSLA"]
sources = ["yahoo.com","businessinsider.com","marketwatch.com"]
for c in companies:
#c = 'NFLX'
    for s in sources:
#s = 'yahoo.com'
        getdata(c,s)
#print(data)
        
```
